```python
import pandas as pd
import glob
import os
import numpy as np
import re
```

```python
df_ipd = pd.read_csv('/home/student/Cloud/Owncloud/SyncVM/CIP project/ljubisavljevic/Data/ipad (1).csv')
df_iph = pd.read_csv('/home/student/Cloud/Owncloud/SyncVM/CIP project/ljubisavljevic/Data/iphone (1).csv')

df1 = pd.concat([df_ipd, df_iph], sort= False)
df1
```

```python
# Rename columnes

df1 = df1.rename(
    columns={'iphone_model' : 'product_model', 
             'iphone_group' : 'product_group'})
df1
```

```python
list(df1.columns)
```

```python
# Removing unnecessary columns

df1 = df1.drop(columns=['web-scraper-order',
                        'web-scraper-start-url',
                        'iphone_model-href',
                        'product_group-href',
                        'iphone_group-href'
                        ])
df1
```

```python
# Overview of product model

list(df1['product_model'].unique())
```

```python
rows_drop = ['iPad vergleichen','iPhone vergleichen']

df1 = df1.drop(df1[df1['product_model'].isin(rows_drop)].index)
                
df1
```

```python
df1.iloc[:,4]
```

```python
# Merging two columns with same name !!DO NOT RUN TWICE!!
df1.iloc[:, 4] = df1.iloc[:, 4].fillna(df1.iloc[:, 6])
df1
```

```python
# Dropping down columns with INDEX   !!DO NOT RUN TWICE!!

df1 = df1.iloc[:, :-1]
df1
```

```python
# Extract model number using regular expressions
df1['model_number'] = df1['model_title'].str.extract(r'(\b[A-Z0-9]{4,}/[A-Z0-9]+\b)').fillna(df1['model_title'].str.extract(r'(\w+/\w+ \w+)'))

# Extract storage using regular expressions
df1['storage'] = df1['model_title'].str.extract(r'(\d+ (?:GB|TB))')

# Extract rating using regular expressions
df1['extracted_rating'] = df1['rating'].str.extract(r'\((\d+(?:\.\d+)?) / \d+\)').fillna(df1['rating'].str.extract(r'\((\d+(?:\.\d+)?) / \d+\)'))

# Print the updated DataFrame
df1
```

```python
# Dropping column with string rating

df1 = df1.drop(columns = 'rating')

df1
```

```python
list(df1.columns)
```

```python
# Changing order of column

df1 = df1.reindex(columns= ['product_group',
                            'product_model',
                            'model_title',
                            'price',
                            'model_number',
                            'storage',
                            'extracted_rating',
                            'ratings_number',])
df1
```

```python
# Renaming column rating

df1 = df1.rename(columns={'extracted_rating' : 'rating'})
df1
```

```python
# Count the number of rows before removing duplicates
original_row_count = len(df1)

# Remove duplicates
df1 = df1.drop_duplicates()

# Count the number of rows after removing duplicates
final_row_count = len(df1)

# Calculate the number of removed duplicates
removed_duplicates = original_row_count - final_row_count

# Print the number of removed duplicates
print(f"Removed {removed_duplicates} duplicate rows.")
```

```python
#fRemoving '€ '

df1['price'] = df1['price'].str.replace('€ ', '').str.replace('.', '').replace(',', '.')

# Remove empty spaces between values in the 'storage' column
df1['storage'] = df1['storage'].str.replace(r'\s+', '', regex=True)

# Remove text after number in 'ratings_number' column
df1['ratings_number'] = df1['ratings_number'].str.replace(r'\D+$', '', regex=True)

# Removing wrongly extracted values 
df1['model_number'] = df1['model_number'].replace(['SSD', 'TB', 'GB'], np.nan, regex=True)

df1
```

```python
df1.to_csv('df1_stage_3.csv', index=False) 
```
