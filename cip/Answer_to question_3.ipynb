# Analysis of dependencies between diffence in prices of Apple products (in Germany and US) and rating with numbers of reviews

```python
import pandas as pd
import numpy as np
```

```python
# reading dtaset from US 
abt = pd.read_csv('../data/abt_rudyka_stage3.csv', sep =',', index_col = 0)
abt.head()
```

```python
# filling NAs
abt['reviews_number'] = abt['reviews_number'].fillna(0)
abt['rating_float'] = abt['rating_float'].fillna(0)
```

```python
#reading dataset from Germany 
cyberport = pd.read_csv('/Users/air/switchdrive/SyncVM/CIP/CIP project/ljubisavljevic/Data/df3_stage_3.csv') 
```

```python
cyberport['ratings_number'] = cyberport['ratings_number'].fillna(0)
cyberport['rating'] = cyberport['rating'].fillna(0)
```

```python
# merging two datagrames
df = pd.merge(abt, cyberport, how='inner', left_on=abt['model_id'].str[:5], right_on=cyberport['model_number'].str[:5])
```

## Calculations of combined data 

```python
df['all_reviews'] = df['reviews_number'] + df['ratings_number']
df['avg_rating'] = (df['reviews_number']*df['rating_float']+df['ratings_number']*df['rating'])/df['all_reviews']
```

```python
eur_dol = 1.0752

df['price_de'] = df['price']*eur_dol
```

```python
df['diff_price'] = (df['price_de']-df['price_cleaned'])/df['price_cleaned']
```

## Modelling with LM OLS

```python
import statsmodels.api as sm
```

```python
# Define the independent variables and the dependent variable
X = df[['all_reviews', 'avg_rating']]
y = df['diff_price']
```

```python
# Add a constant term to the independent variables
X = sm.add_constant(X)

# Fit the linear regression model
model = sm.OLS(y, X)
results = model.fit()

# Print the summary
print(results.summary())
```
